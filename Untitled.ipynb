```python
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import precision_score, recall_score, f1_score
import warnings

# 忽略所有警告
warnings.filterwarnings("ignore")

# 定义词汇表和标签集字典
word_to_idx = {}
tag_to_idx = {}

# 打开数据文件并逐行读取
with open('PeopleDaily199801.txt', 'r', encoding='utf-8') as file:
    for line in file:
        tokens = line.strip().split()  # 去除首尾空白并按空格分割句子
        for token in tokens[1:]:  # 跳过开头的标识符，从第二个词开始处理
            word, tag = token.rsplit('/', 1)
            if word not in word_to_idx:
                word_to_idx[word] = len(word_to_idx) + 1  # 索引从1开始
            if tag not in tag_to_idx:
                tag_to_idx[tag] = len(tag_to_idx) + 1  # 索引从1开始

# 添加特殊符号
special_tokens = ['<PAD>', '<UNK>']
for token in special_tokens:
    if token not in word_to_idx:
        word_to_idx[token] = len(word_to_idx) + 1

# 反向字典
idx_to_word = {v: k for k, v in word_to_idx.items()}
idx_to_tag = {v: k for k, v in tag_to_idx.items()}

# 将每个词语和对应的标签转换为索引形式
sentences = []
labels = []

with open('PeopleDaily199801.txt', 'r', encoding='utf-8') as file:
    for line in file:
        tokens = line.strip().split()
        sentence = []
        label = []
        for token in tokens[1:]:
            word, tag = token.rsplit('/', 1)
            sentence.append(word_to_idx.get(word, word_to_idx['<UNK>']))
            label.append(tag_to_idx[tag])
        sentences.append(sentence)
        labels.append(label)

# 定义数据集类
class POSDataset(Dataset):
    def __init__(self, sentences, labels, max_length):
        self.sentences = sentences
        self.labels = labels
        self.max_length = max_length
    
    def __len__(self):
        return len(self.sentences)
    
    def __getitem__(self, idx):
        sentence = self.sentences[idx]
        padded_sentence = self.pad_or_truncate(sentence, self.max_length)
        label = self.labels[idx]
        padded_label = self.pad_or_truncate(label, self.max_length)
        return torch.tensor(padded_sentence, dtype=torch.long), torch.tensor(padded_label, dtype=torch.long)
    
    def pad_or_truncate(self, seq, max_length):
        if len(seq) < max_length:
            padded_seq = seq + [0] * (max_length - len(seq))  # 使用 0 进行填充
            return padded_seq
        else:
            truncated_seq = seq[:max_length]  # 截断多余部分
            return truncated_seq

# 自定义 collate_fn
def collate_fn(batch):
    sentences, labels = zip(*batch)
    sentences = torch.stack(sentences)
    labels = torch.stack(labels)
    return sentences, labels
        
# 划分训练集和测试集
train_ratio = 0.8

train_sentences = sentences[:int(len(sentences) * train_ratio)]
train_labels = labels[:int(len(labels) * train_ratio)]
test_sentences = sentences[int(len(sentences) * train_ratio):]
test_labels = labels[int(len(labels) * train_ratio):]

max_length = 50  # 假设句子的最大长度为 50
train_dataset = POSDataset(train_sentences, train_labels, max_length)
test_dataset = POSDataset(test_sentences, test_labels, max_length)

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True, collate_fn=collate_fn)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False, collate_fn=collate_fn)

class RNNTagger(nn.Module):
    def __init__(self, vocab_size, tagset_size, embedding_dim=128, hidden_dim=128):
        super(RNNTagger, self).__init__()
        self.embedding = nn.Embedding(vocab_size, embedding_dim)
        self.rnn = nn.RNN(embedding_dim, hidden_dim, batch_first=True)
        self.fc = nn.Linear(hidden_dim, tagset_size)
        self.tagset_size = tagset_size  # 添加这一行以修正 forward 函数中的错误
    
    def forward(self, x):
        embeds = self.embedding(x)
        rnn_out, _ = self.rnn(embeds)

        # Flatten rnn_out to (batch_size * seq_len, hidden_dim)
        rnn_out = rnn_out.contiguous().view(-1, rnn_out.shape[2])

        tag_space = self.fc(rnn_out)

        # Reshape tag_space to (batch_size, seq_len, tagset_size)
        tag_space = tag_space.view(x.size(0), -1, self.tagset_size)

        return tag_space


# 设置设备
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# 超参数
vocab_size = len(word_to_idx) + 1  # 包含padding的索引
tagset_size = len(tag_to_idx) + 1  # 包含padding的索引
embedding_dim = 128
hidden_dim = 128
num_epochs = 100
learning_rate = 0.001

# 模型、损失函数和优化器
model = RNNTagger(vocab_size, tagset_size, embedding_dim, hidden_dim).to(device)
criterion = nn.CrossEntropyLoss(ignore_index=word_to_idx['<PAD>'])
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)

# 训练与评估
train_losses = []
test_losses = []
accuracies = []
precisions = []
recalls = []
f1_scores = []

def calculate_metrics(outputs, labels):
    _, preds = torch.max(outputs, dim=1)
    accuracy = (preds == labels).float().mean().item()
    precision = precision_score(labels.cpu(), preds.cpu(), average='macro')
    recall = recall_score(labels.cpu(), preds.cpu(), average='macro')
    f1 = f1_score(labels.cpu(), preds.cpu(), average='macro')
    return accuracy, precision, recall, f1

for epoch in range(num_epochs):
    model.train()
    total_loss = 0
    for sentences, labels in train_loader:
        sentences, labels = sentences.to(device), labels.to(device)
        optimizer.zero_grad()
        outputs = model(sentences)
        
        # Flatten the predictions and targets
        outputs = outputs.view(-1, outputs.shape[2])
        labels = labels.view(-1)
        
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    
    train_loss = total_loss / len(train_loader)
    train_losses.append(train_loss)
    
    # 在验证集上进行评估
    model.eval()
    test_loss = 0
    all_labels = []
    all_outputs = []
    with torch.no_grad():
        for sentences, labels in test_loader:
            sentences, labels = sentences.to(device), labels.to(device)
            outputs = model(sentences)
            
            # Flatten the predictions and targets
            outputs = outputs.view(-1, outputs.shape[2])
            labels = labels.view(-1)
            
            loss = criterion(outputs, labels)
            test_loss += loss.item()
            
            all_labels.append(labels)
            all_outputs.append(outputs)
    
    test_loss /= len(test_loader)
    test_losses.append(test_loss)
    
    all_labels = torch.cat(all_labels)
    all_outputs = torch.cat(all_outputs)
    
    accuracy, precision, recall, f1 = calculate_metrics(all_outputs, all_labels)
    accuracies.append(accuracy)
    precisions.append(precision)
    recalls.append(recall)
    f1_scores.append(f1)
    
    print(f'Epoch {epoch+1}/{num_epochs}, Loss: {train_loss:.4f}, '
          f'Accuracy: {accuracy:.4f}, Precision: {precision:.4f}, Recall: {recall:.4f}, F1 Score: {f1:.4f}')

# 绘制图表
epochs = range(1, num_epochs + 1)

plt.figure(figsize=(12, 8))
plt.subplot(2, 2, 1)
plt.plot(epochs, train_losses, 'b', label='Train Loss')
# plt.plot(epochs, test_losses, 'r', label='test Loss')
plt.title('Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.subplot(2, 2, 2)
plt.plot(epochs, accuracies, 'b', label='Accuracy')
plt.title('Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

plt.subplot(2, 2, 3)
plt.plot(epochs, precisions, 'b', label='Precision')
plt.title('Precision')
plt.xlabel('Epochs')
plt.ylabel('Precision')
plt.legend()

plt.subplot(2, 2, 4)
plt.plot(epochs, recalls, 'b', label='Recall')
plt.plot(epochs, f1_scores, 'r', label='F1 Score')
plt.title('Recall & F1 Score')
plt.xlabel('Epochs')
plt.ylabel('Score')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# 预测函数
def predict(model, sentence, word_to_idx, idx_to_tag, max_length, device):
    model.eval()
    with torch.no_grad():
        indexed_sentence = [word_to_idx.get(word, word_to_idx['<UNK>']) for word in sentence]
        if len(indexed_sentence) < max_length:
            padded_sentence = indexed_sentence + [word_to_idx["<PAD>"]] * (max_length - len(indexed_sentence))
        else:
            padded_sentence = indexed_sentence[:max_length]
        input_tensor = torch.tensor([padded_sentence], dtype=torch.long).to(device)
        outputs = model(input_tensor)
        outputs = outputs.view(-1, outputs.shape[2])
        _, predicted_indices = torch.max(outputs, 1)

        # 转换为标签，同时忽略填充符号的索引
        predicted_labels = [idx_to_tag[idx.item()] for idx in predicted_indices if idx.item() != word_to_idx['<PAD>']]

        return predicted_labels[:len(sentence)]

# 示例句子
example_sentence = ["今天", "中华人民共和国", "成立", "啦"]

# 进行预测
predicted_tags = predict(model, example_sentence, word_to_idx, idx_to_tag, max_length, device)
print(predicted_tags)
```

```python
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import precision_score, recall_score, f1_score
import warnings

# 忽略所有警告
warnings.filterwarnings("ignore")

# 定义词汇表和标签集字典
word_to_idx = {}
tag_to_idx = {}

# 打开数据文件并逐行读取
with open('PeopleDaily199801.txt', 'r', encoding='utf-8') as file:
    for line in file:
        tokens = line.strip().split()  # 去除首尾空白并按空格分割句子
        for token in tokens[1:]:  # 跳过开头的标识符，从第二个词开始处理
            word, tag = token.rsplit('/', 1)
            if word not in word_to_idx:
                word_to_idx[word] = len(word_to_idx) + 1  # 索引从1开始
            if tag not in tag_to_idx:
                tag_to_idx[tag] = len(tag_to_idx) + 1  # 索引从1开始

# 添加特殊符号
special_tokens = ['<PAD>', '<UNK>']
for token in special_tokens:
    if token not in word_to_idx:
        word_to_idx[token] = len(word_to_idx) + 1

# 反向字典
idx_to_word = {v: k for k, v in word_to_idx.items()}
idx_to_tag = {v: k for k, v in tag_to_idx.items()}

# 将每个词语和对应的标签转换为索引形式
sentences = []
labels = []

with open('PeopleDaily199801.txt', 'r', encoding='utf-8') as file:
    for line in file:
        tokens = line.strip().split()
        sentence = []
        label = []
        for token in tokens[1:]:
            word, tag = token.rsplit('/', 1)
            sentence.append(word_to_idx.get(word, word_to_idx['<UNK>']))
            label.append(tag_to_idx[tag])
        sentences.append(sentence)
        labels.append(label)

# 定义数据集类
class POSDataset(Dataset):
    def __init__(self, sentences, labels, max_length):
        self.sentences = sentences
        self.labels = labels
        self.max_length = max_length
    
    def __len__(self):
        return len(self.sentences)
    
    def __getitem__(self, idx):
        sentence = self.sentences[idx]
        padded_sentence = self.pad_or_truncate(sentence, self.max_length)
        label = self.labels[idx]
        padded_label = self.pad_or_truncate(label, self.max_length)
        return torch.tensor(padded_sentence, dtype=torch.long), torch.tensor(padded_label, dtype=torch.long)
    
    def pad_or_truncate(self, seq, max_length):
        if len(seq) < max_length:
            padded_seq = seq + [0] * (max_length - len(seq))  # 使用 0 进行填充
            return padded_seq
        else:
            truncated_seq = seq[:max_length]  # 截断多余部分
            return truncated_seq

# 自定义 collate_fn
def collate_fn(batch):
    sentences, labels = zip(*batch)
    sentences = torch.stack(sentences)
    labels = torch.stack(labels)
    return sentences, labels
        
# 划分训练集和测试集
train_ratio = 0.8

train_sentences = sentences[:int(len(sentences) * train_ratio)]
train_labels = labels[:int(len(labels) * train_ratio)]
test_sentences = sentences[int(len(sentences) * train_ratio):]
test_labels = labels[int(len(labels) * train_ratio):]

max_length = 50  # 假设句子的最大长度为 50
train_dataset = POSDataset(train_sentences, train_labels, max_length)
test_dataset = POSDataset(test_sentences, test_labels, max_length)

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True, collate_fn=collate_fn)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False, collate_fn=collate_fn)

# 定义基于Transformer的模型
class TransformerTagger(nn.Module):
    def __init__(self, vocab_size, tagset_size, embedding_dim=128, num_heads=8, num_layers=2, hidden_dim=256):
        super(TransformerTagger, self).__init__()
        self.embedding = nn.Embedding(vocab_size, embedding_dim)
        self.pos_encoder = nn.TransformerEncoderLayer(d_model=embedding_dim, nhead=num_heads)
        self.transformer = nn.TransformerEncoder(self.pos_encoder, num_layers=num_layers)
        self.fc = nn.Linear(embedding_dim, tagset_size)
        self.tagset_size = tagset_size

    def forward(self, x):
        embeds = self.embedding(x) * math.sqrt(embedding_dim)
        embeds = self.add_positional_encoding(embeds)
        transformer_out = self.transformer(embeds)
        
        # Flatten transformer_out to (batch_size * seq_len, embedding_dim)
        transformer_out = transformer_out.contiguous().view(-1, transformer_out.shape[2])

        tag_space = self.fc(transformer_out)

        # Reshape tag_space to (batch_size, seq_len, tagset_size)
        tag_space = tag_space.view(x.size(0), -1, self.tagset_size)

        return tag_space

    def add_positional_encoding(self, x):
        # Adding positional encoding
        pe = torch.zeros(x.size(1), x.size(2)).to(x.device)
        position = torch.arange(0, x.size(1), dtype=torch.float).unsqueeze(1).to(x.device)
        div_term = torch.exp(torch.arange(0, x.size(2), 2).float() * (-math.log(10000.0) / x.size(2))).to(x.device)
        pe[:, 0::2] = torch.sin(position * div_term)
        pe[:, 1::2] = torch.cos(position * div_term)
        pe = pe.unsqueeze(0)
        return x + pe

import math

# 设置设备
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# 超参数
vocab_size = len(word_to_idx) + 1  # 包含padding的索引
tagset_size = len(tag_to_idx) + 1  # 包含padding的索引
embedding_dim = 128
hidden_dim = 128
num_epochs = 100
learning_rate = 0.001

# 模型、损失函数和优化器
model = TransformerTagger(vocab_size, tagset_size, embedding_dim, hidden_dim).to(device)
criterion = nn.CrossEntropyLoss(ignore_index=word_to_idx['<PAD>'])
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)

# 训练与评估
train_losses = []
test_losses = []
accuracies = []
precisions = []
recalls = []
f1_scores = []

def calculate_metrics(outputs, labels):
    _, preds = torch.max(outputs, dim=1)
    accuracy = (preds == labels).float().mean().item()
    precision = precision_score(labels.cpu(), preds.cpu(), average='macro')
    recall = recall_score(labels.cpu(), preds.cpu(), average='macro')
    f1 = f1_score(labels.cpu(), preds.cpu(), average='macro')
    return accuracy, precision, recall, f1

for epoch in range(num_epochs):
    model.train()
    total_loss = 0
    for sentences, labels in train_loader:
        sentences, labels = sentences.to(device), labels.to(device)
        optimizer.zero_grad()
        outputs = model(sentences)
        
        # Flatten the predictions and targets
        outputs = outputs.view(-1, outputs.shape[2])
        labels = labels.view(-1)
        
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    
    train_loss = total_loss / len(train_loader)
    train_losses.append(train_loss)
    
    # 在验证集上进行评估
    model.eval()
    test_loss = 0
    all_labels = []
    all_outputs = []
    with torch.no_grad():
        for sentences, labels in test_loader:
            sentences, labels = sentences.to(device), labels.to(device)
            outputs = model(sentences)
            
            # Flatten the predictions and targets
            outputs = outputs.view(-1, outputs.shape[2])
            labels = labels.view(-1)
            
            loss = criterion(outputs, labels)
            test_loss += loss.item()
            
            all_labels.append(labels)
            all_outputs.append(outputs)
    
    test_loss /= len(test_loader)
    test_losses.append(test_loss)
    
    all_labels = torch.cat(all_labels)
    all_outputs = torch.cat(all_outputs)
    
    accuracy, precision, recall, f1 = calculate_metrics(all_outputs, all_labels)
    accuracies.append(accuracy)
    precisions.append(precision)
    recalls.append(recall)
    f1_scores.append(f1)
    
    print(f'Epoch {epoch+1}/{num_epochs}, Loss: {train_loss:.4f}, '
          f'Accuracy: {accuracy:.4f}, Precision: {precision:.4f}, Recall: {recall:.4f}, F1 Score: {f1:.4f}')

# 绘制图表
epochs = range(1, num_epochs + 1)

plt.figure(figsize=(12, 8))
plt.subplot(2, 2, 1)
plt.plot(epochs, train_losses, 'b', label='Train Loss')
# plt.plot(epochs, test_losses, 'r', label='test Loss')
plt.title('Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.subplot(2, 2, 2)
plt.plot(epochs, accuracies, 'b', label='Accuracy')
plt.title('Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

plt.subplot(2, 2, 3)
plt.plot(epochs, precisions, 'b', label='Precision')
plt.title('Precision')
plt.xlabel('Epochs')
plt.ylabel('Precision')
plt.legend()

plt.subplot(2, 2, 4)
plt.plot(epochs, recalls, 'b', label='Recall')
plt.plot(epochs, f1_scores, 'r', label='F1 Score')
plt.title('Recall & F1 Score')
plt.xlabel('Epochs')
plt.ylabel('Score')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import precision_score, recall_score, f1_score
import warnings

# 忽略所有警告
warnings.filterwarnings("ignore")

# 定义词汇表和标签集字典
word_to_idx = {}
tag_to_idx = {}

base_path = os.listdir('20140102')
for i in base_path:
    for j in os.listdir('20140102/' + i):
        # 打开数据文件并逐行读取
        with open('20140102/' + i + '/' + j, 'r', encoding='utf-8') as file:
            for line in file:
                tokens = line.strip().split()  # 去除首尾空白并按空格分割句子
                for token in tokens:  # 跳过开头的标识符，从第二个词开始处理
                    word, tag = token.rsplit('/', 1)
                    if word not in word_to_idx:
                        word_to_idx[word] = len(word_to_idx) + 1  # 索引从1开始
                    if tag not in tag_to_idx:
                        tag_to_idx[tag] = len(tag_to_idx) + 1  # 索引从1开始

# 添加特殊符号
special_tokens = ['<PAD>', '<UNK>']
for token in special_tokens:
    if token not in word_to_idx:
        word_to_idx[token] = len(word_to_idx) + 1

# 反向字典
idx_to_word = {v: k for k, v in word_to_idx.items()}
idx_to_tag = {v: k for k, v in tag_to_idx.items()}

# 将每个词语和对应的标签转换为索引形式
sentences = []
labels = []

for i in base_path:
    for j in os.listdir('20140102/' + i):
        with open('20140102/' + i + '/' + j, 'r', encoding='utf-8') as file:
            for line in file:
                tokens = line.strip().split()
                sentence = []
                label = []
                for token in tokens[1:]:
                    word, tag = token.rsplit('/', 1)
                    sentence.append(word_to_idx.get(word, word_to_idx['<UNK>']))
                    label.append(tag_to_idx[tag])
                sentences.append(sentence)
                labels.append(label)

# 定义数据集类
class POSDataset(Dataset):
    def __init__(self, sentences, labels, max_length):
        self.sentences = sentences
        self.labels = labels
        self.max_length = max_length
    
    def __len__(self):
        return len(self.sentences)
    
    def __getitem__(self, idx):
        sentence = self.sentences[idx]
        padded_sentence = self.pad_or_truncate(sentence, self.max_length)
        label = self.labels[idx]
        padded_label = self.pad_or_truncate(label, self.max_length)
        return torch.tensor(padded_sentence, dtype=torch.long), torch.tensor(padded_label, dtype=torch.long)
    
    def pad_or_truncate(self, seq, max_length):
        if len(seq) < max_length:
            padded_seq = seq + [0] * (max_length - len(seq))  # 使用 0 进行填充
            return padded_seq
        else:
            truncated_seq = seq[:max_length]  # 截断多余部分
            return truncated_seq

# 自定义 collate_fn
def collate_fn(batch):
    sentences, labels = zip(*batch)
    sentences = torch.stack(sentences)
    labels = torch.stack(labels)
    return sentences, labels
        
# 划分训练集和测试集
train_ratio = 0.8

train_sentences = sentences[:int(len(sentences) * train_ratio)]
train_labels = labels[:int(len(labels) * train_ratio)]
test_sentences = sentences[int(len(sentences) * train_ratio):]
test_labels = labels[int(len(labels) * train_ratio):]

max_length = 50  # 假设句子的最大长度为 50
train_dataset = POSDataset(train_sentences, train_labels, max_length)
test_dataset = POSDataset(test_sentences, test_labels, max_length)

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True, collate_fn=collate_fn)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False, collate_fn=collate_fn)

# 定义LSTM模型
class LSTMTagger(nn.Module):
    def __init__(self, vocab_size, tagset_size, embedding_dim=128, hidden_dim=128):
        super(LSTMTagger, self).__init__()
        self.embedding = nn.Embedding(vocab_size, embedding_dim)
        self.lstm = nn.LSTM(embedding_dim, hidden_dim, batch_first=True)
        self.fc = nn.Linear(hidden_dim, tagset_size)
        self.tagset_size = tagset_size
    
    def forward(self, x):
        embeds = self.embedding(x)
        lstm_out, _ = self.lstm(embeds)

        # Flatten lstm_out to (batch_size * seq_len, hidden_dim)
        lstm_out = lstm_out.contiguous().view(-1, lstm_out.shape[2])

        tag_space = self.fc(lstm_out)

        # Reshape tag_space to (batch_size, seq_len, tagset_size)
        tag_space = tag_space.view(x.size(0), -1, self.tagset_size)

        return tag_space


# 设置设备
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# 超参数
vocab_size = len(word_to_idx) + 1  # 包含padding的索引
tagset_size = len(tag_to_idx) + 1  # 包含padding的索引
embedding_dim = 128
hidden_dim = 128
num_epochs = 100
learning_rate = 0.001

# 模型、损失函数和优化器
model = LSTMTagger(vocab_size, tagset_size, embedding_dim, hidden_dim).to(device)
criterion = nn.CrossEntropyLoss(ignore_index=word_to_idx['<PAD>'])
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)

# 训练与评估
train_losses = []
test_losses = []
accuracies = []
precisions = []
recalls = []
f1_scores = []

def calculate_metrics(outputs, labels):
    _, preds = torch.max(outputs, dim=1)
    accuracy = (preds == labels).float().mean().item()
    precision = precision_score(labels.cpu(), preds.cpu(), average='macro')
    recall = recall_score(labels.cpu(), preds.cpu(), average='macro')
    f1 = f1_score(labels.cpu(), preds.cpu(), average='macro')
    return accuracy, precision, recall, f1

for epoch in range(num_epochs):
    model.train()
    total_loss = 0
    for sentences, labels in train_loader:
        sentences, labels = sentences.to(device), labels.to(device)
        optimizer.zero_grad()
        outputs = model(sentences)
        
        # Flatten the predictions and targets
        outputs = outputs.view(-1, outputs.shape[2])
        labels = labels.view(-1)
        
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    
    train_loss = total_loss / len(train_loader)
    train_losses.append(train_loss)
    
    # 在验证集上进行评估
    model.eval()
    test_loss = 0
    all_labels = []
    all_outputs = []
    with torch.no_grad():
        for sentences, labels in test_loader:
            sentences, labels = sentences.to(device), labels.to(device)
            outputs = model(sentences)
            
            # Flatten the predictions and targets
            outputs = outputs.view(-1, outputs.shape[2])
            labels = labels.view(-1)
            
            loss = criterion(outputs, labels)
            test_loss += loss.item()
            
            all_labels.append(labels)
            all_outputs.append(outputs)
    
    test_loss /= len(test_loader)
    test_losses.append(test_loss)
    
    all_labels = torch.cat(all_labels)
    all_outputs = torch.cat(all_outputs)
    
    accuracy, precision, recall, f1 = calculate_metrics(all_outputs, all_labels)
    accuracies.append(accuracy)
    precisions.append(precision)
    recalls.append(recall)
    f1_scores.append(f1)
    
    print(f'Epoch {epoch+1}/{num_epochs}, Loss: {train_loss:.4f}, '
          f'Accuracy: {accuracy:.4f}, Precision: {precision:.4f}, Recall: {recall:.4f}, F1 Score: {f1:.4f}')

# 绘制图表
epochs = range(1, num_epochs + 1)

plt.figure(figsize=(12, 8))
plt.subplot(2, 2, 1)
plt.plot(epochs, train_losses, 'b', label='Train Loss')
# plt.plot(epochs, test_losses, 'r', label='test Loss')
plt.title('Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.subplot(2, 2, 2)
plt.plot(epochs, accuracies, 'b', label='Accuracy')
plt.title('Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

plt.subplot(2, 2, 3)
plt.plot(epochs, precisions, 'b', label='Precision')
plt.title('Precision')
plt.xlabel('Epochs')
plt.ylabel('Precision')
plt.legend()

plt.subplot(2, 2, 4)
plt.plot(epochs, recalls, 'b', label='Recall')
plt.plot(epochs, f1_scores, 'r', label='F1 Score')
plt.title('Recall & F1 Score')
plt.xlabel('Epochs')
plt.ylabel('Score')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import precision_score, recall_score, f1_score
import warnings

# 忽略所有警告
warnings.filterwarnings("ignore")

# 定义词汇表和标签集字典
word_to_idx = {}
tag_to_idx = {}

base_path = os.listdir('20140102')
for i in base_path:
    for j in os.listdir('20140102/' + i):
        # 打开数据文件并逐行读取
        with open('20140102/' + i + '/' + j, 'r', encoding='utf-8') as file:
            for line in file:
                tokens = line.strip().split()  # 去除首尾空白并按空格分割句子
                for token in tokens:  # 跳过开头的标识符，从第二个词开始处理
                    word, tag = token.rsplit('/', 1)
                    if word not in word_to_idx:
                        word_to_idx[word] = len(word_to_idx) + 1  # 索引从1开始
                    if tag not in tag_to_idx:
                        tag_to_idx[tag] = len(tag_to_idx) + 1  # 索引从1开始

# 添加特殊符号
special_tokens = ['<PAD>', '<UNK>']
for token in special_tokens:
    if token not in word_to_idx:
        word_to_idx[token] = len(word_to_idx) + 1

# 反向字典
idx_to_word = {v: k for k, v in word_to_idx.items()}
idx_to_tag = {v: k for k, v in tag_to_idx.items()}

# 将每个词语和对应的标签转换为索引形式
sentences = []
labels = []

for i in base_path:
    for j in os.listdir('20140102/' + i):
        with open('20140102/' + i + '/' + j, 'r', encoding='utf-8') as file:
            for line in file:
                tokens = line.strip().split()
                sentence = []
                label = []
                for token in tokens[1:]:
                    word, tag = token.rsplit('/', 1)
                    sentence.append(word_to_idx.get(word, word_to_idx['<UNK>']))
                    label.append(tag_to_idx[tag])
                sentences.append(sentence)
                labels.append(label)

# 定义数据集类
class POSDataset(Dataset):
    def __init__(self, sentences, labels, max_length):
        self.sentences = sentences
        self.labels = labels
        self.max_length = max_length
    
    def __len__(self):
        return len(self.sentences)
    
    def __getitem__(self, idx):
        sentence = self.sentences[idx]
        padded_sentence = self.pad_or_truncate(sentence, self.max_length)
        label = self.labels[idx]
        padded_label = self.pad_or_truncate(label, self.max_length)
        return torch.tensor(padded_sentence, dtype=torch.long), torch.tensor(padded_label, dtype=torch.long)
    
    def pad_or_truncate(self, seq, max_length):
        if len(seq) < max_length:
            padded_seq = seq + [0] * (max_length - len(seq))  # 使用 0 进行填充
            return padded_seq
        else:
            truncated_seq = seq[:max_length]  # 截断多余部分
            return truncated_seq

# 自定义 collate_fn
def collate_fn(batch):
    sentences, labels = zip(*batch)
    sentences = torch.stack(sentences)
    labels = torch.stack(labels)
    return sentences, labels
        
# 划分训练集和测试集
train_ratio = 0.8

train_sentences = sentences[:int(len(sentences) * train_ratio)]
train_labels = labels[:int(len(labels) * train_ratio)]
test_sentences = sentences[int(len(sentences) * train_ratio):]
test_labels = labels[int(len(labels) * train_ratio):]

max_length = 50  # 假设句子的最大长度为 50
train_dataset = POSDataset(train_sentences, train_labels, max_length)
test_dataset = POSDataset(test_sentences, test_labels, max_length)

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True, collate_fn=collate_fn)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False, collate_fn=collate_fn)

# 定义基于Transformer的模型
class TransformerTagger(nn.Module):
    def __init__(self, vocab_size, tagset_size, embedding_dim=128, num_heads=8, num_layers=2, hidden_dim=256):
        super(TransformerTagger, self).__init__()
        self.embedding = nn.Embedding(vocab_size, embedding_dim)
        self.pos_encoder = nn.TransformerEncoderLayer(d_model=embedding_dim, nhead=num_heads)
        self.transformer = nn.TransformerEncoder(self.pos_encoder, num_layers=num_layers)
        self.fc = nn.Linear(embedding_dim, tagset_size)
        self.tagset_size = tagset_size

    def forward(self, x):
        embeds = self.embedding(x) * math.sqrt(embedding_dim)
        embeds = self.add_positional_encoding(embeds)
        transformer_out = self.transformer(embeds)
        
        # Flatten transformer_out to (batch_size * seq_len, embedding_dim)
        transformer_out = transformer_out.contiguous().view(-1, transformer_out.shape[2])

        tag_space = self.fc(transformer_out)

        # Reshape tag_space to (batch_size, seq_len, tagset_size)
        tag_space = tag_space.view(x.size(0), -1, self.tagset_size)

        return tag_space

    def add_positional_encoding(self, x):
        # Adding positional encoding
        pe = torch.zeros(x.size(1), x.size(2)).to(x.device)
        position = torch.arange(0, x.size(1), dtype=torch.float).unsqueeze(1).to(x.device)
        div_term = torch.exp(torch.arange(0, x.size(2), 2).float() * (-math.log(10000.0) / x.size(2))).to(x.device)
        pe[:, 0::2] = torch.sin(position * div_term)
        pe[:, 1::2] = torch.cos(position * div_term)
        pe = pe.unsqueeze(0)
        return x + pe

import math

# 设置设备
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# 超参数
vocab_size = len(word_to_idx) + 1  # 包含padding的索引
tagset_size = len(tag_to_idx) + 1  # 包含padding的索引
embedding_dim = 128
hidden_dim = 128
num_epochs = 100
learning_rate = 0.001

# 模型、损失函数和优化器
model = TransformerTagger(vocab_size, tagset_size, embedding_dim, hidden_dim).to(device)
criterion = nn.CrossEntropyLoss(ignore_index=word_to_idx['<PAD>'])
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)

# 训练与评估
train_losses = []
test_losses = []
accuracies = []
precisions = []
recalls = []
f1_scores = []

def calculate_metrics(outputs, labels):
    _, preds = torch.max(outputs, dim=1)
    accuracy = (preds == labels).float().mean().item()
    precision = precision_score(labels.cpu(), preds.cpu(), average='macro')
    recall = recall_score(labels.cpu(), preds.cpu(), average='macro')
    f1 = f1_score(labels.cpu(), preds.cpu(), average='macro')
    return accuracy, precision, recall, f1

for epoch in range(num_epochs):
    model.train()
    total_loss = 0
    for sentences, labels in train_loader:
        sentences, labels = sentences.to(device), labels.to(device)
        optimizer.zero_grad()
        outputs = model(sentences)
        
        # Flatten the predictions and targets
        outputs = outputs.view(-1, outputs.shape[2])
        labels = labels.view(-1)
        
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    
    train_loss = total_loss / len(train_loader)
    train_losses.append(train_loss)
    
    # 在验证集上进行评估
    model.eval()
    test_loss = 0
    all_labels = []
    all_outputs = []
    with torch.no_grad():
        for sentences, labels in test_loader:
            sentences, labels = sentences.to(device), labels.to(device)
            outputs = model(sentences)
            
            # Flatten the predictions and targets
            outputs = outputs.view(-1, outputs.shape[2])
            labels = labels.view(-1)
            
            loss = criterion(outputs, labels)
            test_loss += loss.item()
            
            all_labels.append(labels)
            all_outputs.append(outputs)
    
    test_loss /= len(test_loader)
    test_losses.append(test_loss)
    
    all_labels = torch.cat(all_labels)
    all_outputs = torch.cat(all_outputs)
    
    accuracy, precision, recall, f1 = calculate_metrics(all_outputs, all_labels)
    accuracies.append(accuracy)
    precisions.append(precision)
    recalls.append(recall)
    f1_scores.append(f1)
    
    print(f'Epoch {epoch+1}/{num_epochs}, Loss: {train_loss:.4f}, '
          f'Accuracy: {accuracy:.4f}, Precision: {precision:.4f}, Recall: {recall:.4f}, F1 Score: {f1:.4f}')

# 绘制图表
epochs = range(1, num_epochs + 1)

plt.figure(figsize=(12, 8))
plt.subplot(2, 2, 1)
plt.plot(epochs, train_losses, 'b', label='Train Loss')
# plt.plot(epochs, test_losses, 'r', label='test Loss')
plt.title('Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.subplot(2, 2, 2)
plt.plot(epochs, accuracies, 'b', label='Accuracy')
plt.title('Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

plt.subplot(2, 2, 3)
plt.plot(epochs, precisions, 'b', label='Precision')
plt.title('Precision')
plt.xlabel('Epochs')
plt.ylabel('Precision')
plt.legend()

plt.subplot(2, 2, 4)
plt.plot(epochs, recalls, 'b', label='Recall')
plt.plot(epochs, f1_scores, 'r', label='F1 Score')
plt.title('Recall & F1 Score')
plt.xlabel('Epochs')
plt.ylabel('Score')
plt.legend()

plt.tight_layout()
plt.show()
```
